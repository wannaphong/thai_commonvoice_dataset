```python
import glob
import os
import pathlib
```

```python
import pandas as pd
```

```python
root_dir=os.path.join(".","cv-corpus-8.0-2022-01-19","th")
```

```python
pathlib.Path(os.path.join(root_dir,"wavs")).mkdir(parents=True, exist_ok=True) 
```

```python
validated = os.path.join(root_dir,"validated.tsv")
```

```python
df = pd.read_csv(validated,sep="\t")
```

```python
def get_mp3_path(filename):
    return os.path.join(root_dir,"clips",filename)
```

```python
def get_wav_path(filename):
    return os.path.join(root_dir,"wavs",filename.replace('mp3','wav'))
```

```python
def run(item):
    mp3_path = get_mp3_path(item)
    sr = get_wav_path(item)
    os.system("sox {} -r 16000 -c 1 -b 16 {}".format(mp3_path,sr))
```
